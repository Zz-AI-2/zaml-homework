```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

# 5 sampels


```python
x_5samp=np.random.rand(1 , 5)*60
w_5samp=np.random.rand(1 , 1)*60
b_5samp=np.random.rand(1 , 1)*60
y_5samp=w_5samp*x_5samp+b_5samp
y_5samp=y_5samp+np.random.normal(scale=3, size=y_5samp.shape)

x_5samp = x_5samp.reshape(-1, 1)
y_5samp = y_5samp.reshape(-1, 1)

degree=[1 , 4 , 16 ]
color={ 1 : 'b--' , 4 : 'r--' , 16 : 'y--' }

score_5samp=[]
j=1

for i in degree:
    poly = PolynomialFeatures(degree=i)
    x_poly=poly.fit_transform(x_5samp)

    lr_model_5samp = LinearRegression(normalize=True)
    lr_model_5samp.fit(x_poly, y_5samp)

    score_5samp.append(lr_model_5samp.score(x_poly , y_5samp))
    
    x_line=np.arange(-10 , 60 ,0.1).reshape(-1,1)
    x_line_poly=poly.fit_transform(x_line)
    y_line=lr_model_5samp.predict(x_line_poly)
    degree="degree"+ str(i)
      
    plt.plot(x_5samp, y_5samp , 'ko')
    plt.plot(x_line , y_line , color[i] , label=degree)
    plt.legend()

score_5={"score 5_samp 1 degree:" : score_5samp[0] , "score 5_samp 4 degree :" : score_5samp[1] , 
             "score 5_samp 16 degree :" : score_5samp[2]}    
print(score_5samp)   
```
